## Question 2 (10 Points)
Write python code to extract these features you discussed above. You can collect a few sample text data for the feature extraction.

```python
!pip install scikit-learn nltk spacy vaderSentiment
!pip python -m spacy download en_core_web_sm
import nltk
import spacy
from sklearn.feature_extraction.text import CountVectorizer, TfidfVectorizer
from vaderSentiment.vaderSentiment import SentimentIntensityAnalyzer
from nltk import ngrams
from collections import Counter

# Sample text data
texts = [
    "This product is absolutely amazing! I've never seen anything like it.",
    "Terrible experience. The quality was poor and it broke after one use.",
    "It's okay, but not worth the price. I expected something better."
]

# Initialize necessary tools
nltk.download('punkt')
nltk.download('averaged_perceptron_tagger')
nlp = spacy.load('en_core_web_sm')
sentiment_analyzer = SentimentIntensityAnalyzer()

# 1. Bag of Words (BoW) Features
vectorizer_bow = CountVectorizer()
bow_features = vectorizer_bow.fit_transform(texts).toarray()
print("Bag of Words (BoW) Features:")
print(bow_features)

# 2. TF-IDF Features
vectorizer_tfidf = TfidfVectorizer()
tfidf_features = vectorizer_tfidf.fit_transform(texts).toarray()
print("\nTF-IDF Features:")
print(tfidf_features)

# 3. Sentiment Lexicon-Based Features (Using VADER Sentiment)
sentiment_scores = []
for text in texts:
    scores = sentiment_analyzer.polarity_scores(text)
    sentiment_scores.append([scores['neg'], scores['neu'], scores['pos'], scores['compound']])

print("\nSentiment Lexicon Features (neg, neu, pos, compound):")
print(sentiment_scores)

# 4. N-gram Features (Bi-grams)
n_gram_vectorizer = CountVectorizer(ngram_range=(2, 2))
n_gram_features = n_gram_vectorizer.fit_transform(texts).toarray()
print("\nBi-gram Features:")
print(n_gram_features)

# 5. Part-of-Speech (POS) Tagging Features
def extract_pos_features(text):
    doc = nlp(text)
    pos_counts = Counter([token.pos_ for token in doc])
    return [pos_counts.get('ADJ', 0), pos_counts.get('ADV', 0), pos_counts.get('NOUN', 0), pos_counts.get('VERB', 0)]

pos_features = [extract_pos_features(text) for text in texts]
print("\nPOS Tagging Features (ADJ, ADV, NOUN, VERB counts):")
print(pos_features)

# Additional Features

# a. Length of the review
review_lengths = [len(text.split()) for text in texts]
print("\nLength of Reviews (in words):")
print(review_lengths)

# b. Punctuation Count Features (Exclamation and Question marks)
def extract_punctuation_features(text):
    exclamation_count = text.count('!')
    question_count = text.count('?')
    return [exclamation_count, question_count]

punctuation_features = [extract_punctuation_features(text) for text in texts]
print("\nPunctuation Features (Exclamation marks, Question marks):")
print(punctuation_features)
```

## Question 3 (10 points):
Use any of the feature selection methods mentioned in this paper "Deng, X., Li, Y., Weng, J., & Zhang, J. (2019). Feature selection for text classification: A review. Multimedia Tools & Applications, 78(3)."

Select the most important features you extracted above, rank the features based on their importance in the descending order.

```python
# You code here (Please add comments in the code):

import numpy as np
from sklearn.feature_extraction.text import TfidfVectorizer, CountVectorizer
from sklearn.feature_selection import SelectKBest, chi2
from sklearn.preprocessing import LabelEncoder

# Sample text data and labels
texts = [
    "This product is absolutely amazing! I've never seen anything like it.",
    "Terrible experience. The quality was poor and it broke after one use.",
    "It's okay, but not worth the price. I expected something better."
]
labels = ['positive', 'negative', 'neutral']

# Convert labels to numerical format
label_encoder = LabelEncoder()
y = label_encoder.fit_transform(labels)

# 1. Extract features using TF-IDF
vectorizer = TfidfVectorizer()
X_tfidf = vectorizer.fit_transform(texts)

# 2. Feature selection using Chi-square
chi2_selector = SelectKBest(chi2, k='all')  # 'all' to rank all features
X_kbest = chi2_selector.fit_transform(X_tfidf, y)

# 3. Get feature names and their Chi-square scores
feature_names = vectorizer.get_feature_names_out()
chi2_scores = chi2_selector.scores_

# Combine features with their Chi-square scores
feature_ranking = sorted(zip(feature_names, chi2_scores), key=lambda x: x[1], reverse=True)

# Display features ranked by importance
print("Top features ranked by Chi-square score:")
for feature, score in feature_ranking:
    print(f"{feature}: {score}")


```

## Question 4 (10 points):
Write python code to rank the text based on text similarity. Based on the text data you used for question 2, design a query to match the most relevant docments. Please use the BERT model to represent both your query and the text data, then calculate the cosine similarity between the query and each text in your data. Rank the similary with descending order.

```python
# You code here (Please add comments in the code):
import torch
from transformers import BertTokenizer, BertModel
from sklearn.metrics.pairwise import cosine_similarity
import numpy as np

# Load pre-trained BERT model and tokenizer
tokenizer = BertTokenizer.from_pretrained('bert-base-uncased')
model = BertModel.from_pretrained('bert-base-uncased')

# Sample text data
texts = [
    "This product is absolutely amazing! I've never seen anything like it.",
    "Terrible experience. The quality was poor and it broke after one use.",
    "It's okay, but not worth the price. I expected something better."
]

# Define a query to match against the text data
query = "I love the quality of this product, it's fantastic."

# Function to generate BERT embeddings
def get_bert_embedding(text):
    inputs = tokenizer(text, return_tensors='pt', padding=True, truncation=True, max_length=512)
    with torch.no_grad():
        outputs = model(**inputs)
    # Take the embedding of [CLS] token
    return outputs.last_hidden_state[:, 0, :].numpy()

# 1. Get BERT embedding for the query
query_embedding = get_bert_embedding(query)

# 2. Get BERT embeddings for each text
text_embeddings = [get_bert_embedding(text) for text in texts]

# 3. Calculate cosine similarity between query and each document
similarities = [cosine_similarity(query_embedding, text_emb)[0][0] for text_emb in text_embeddings]

# 4. Rank documents based on cosine similarity
ranked_texts = sorted(zip(texts, similarities), key=lambda x: x[1], reverse=True)

# 5. Display ranked texts and their similarity scores
print("Ranked texts based on similarity to the query:")
for idx, (text, score) in enumerate(ranked_texts):
    print(f"Rank {idx + 1}: (Similarity: {score:.4f})")
    print(f"Text: {text}\n")



```
